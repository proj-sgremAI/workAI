```python
import yfinance as yf
import pandas as pd

tickers=pd.read_csv("C:\\Users\\dssgr\\OneDrive\\Desktop\\folder\\tickers.csv")
tickers=tickers['Symbol'].drop_duplicates()
```

```python
df_finance_t=pd.DataFrame()
```

```python
for i in tickers:
    Ticker= yf.Ticker(i)
    df_finance=Ticker.get_financials()
    df_finance.columns = pd.MultiIndex.from_product([[i], df_finance.columns])
    df_finance_t=pd.concat([df_finance_t, df_finance], axis = 1)
    print(df_finance_t)
```

```python
df_cashflow_t=pd.DataFrame()
```
